```python
from pylab import *
```

```python
import pawdata
```

```python
title("Projector functions of H")
plot(paw.radial_grid_values, paw.projector_function["H1"])
plot(paw.radial_grid_values, paw.projector_function["H2"])
plot(paw.radial_grid_values, paw.projector_function["H3"])
xlim([0, 1])
xlabel("Radius [Bohr]")
```

```python
paw = pawdata.PAWData("H.xml.gz")

fig, [a, b] = subplots(2, sharex=True, sharey=True, figsize=[7, 7])
xlim([0, 10])
xlabel("Radius [Bohr]")

a.set_title("Pseudo partial waves")
for state_id in  paw.pseudo_partial_wave:
    a.plot(paw.radial_grid_values, paw.pseudo_partial_wave[state_id], label=state_id)
a.legend()


b.set_title("AE partial waves")
for state_id in  paw.ae_partial_wave:
    b.plot(paw.radial_grid_values, paw.ae_partial_wave[state_id], label=state_id)
b.legend()
```

```python
paw = pawdata.PAWData("Na.xml.gz")

fig, [a, b] = subplots(2, sharex=True, sharey=True, figsize=[7, 7])
xlim([0, 2])
ylim(-10, 5)
xlabel("Radius [Bohr]")

a.set_title("Pseudo partial waves")
for state_id in  paw.pseudo_partial_wave:
    a.plot(paw.radial_grid_values, paw.pseudo_partial_wave[state_id], label=state_id)
a.legend(loc="lower right")


b.set_title("AE partial waves")
for state_id in  paw.ae_partial_wave:
    b.plot(paw.radial_grid_values, paw.ae_partial_wave[state_id], label=state_id)
b.legend(loc="lower right")
```
